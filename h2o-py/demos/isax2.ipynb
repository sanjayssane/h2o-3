```python
df = h2o.create_frame(cols=2, real_fraction=1.0, missing_fraction=0.0)
```

```python
for c in df.columns:
    df[c] = df[c].cumsum()
```

```python
df.as_data_frame(use_pandas=True).plot()
```

```python
dftrans = df.transpose()
```

```python
res = dftrans.isax(num_words=10,max_cardinality=10)
```
